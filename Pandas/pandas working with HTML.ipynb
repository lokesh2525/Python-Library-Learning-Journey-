Writing HTML Tables

```python
import pandas as pd

# Creating a sample DataFrame
data = {'Name': ['John', 'Anna', 'Mike'],
        'Age': [30, 25, 35]}
df = pd.DataFrame(data)

# Writing the DataFrame to an HTML file
df.to_html('output.html', index=False)
```

Writing to an HTML String

```python
import pandas as pd

# Creating a sample DataFrame
data = {'Name': ['John', 'Anna', 'Mike'],
        'Age': [30, 25, 35]}
df = pd.DataFrame(data)

# Getting the HTML representation of the DataFrame
html_string = df.to_html(index=False)
print(html_string)
```

```python
import pandas as pd

# Reading the first table from a webpage
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
tables = pd.read_html(url)

# Displaying the number of tables found
print(f"Number of tables found: {len(tables)}")

# Selecting the first table and displaying its head
df = tables[0]
print(df.head())

# Writing the DataFrame to an HTML file
df.to_html('countries_population.html', index=False)
```
