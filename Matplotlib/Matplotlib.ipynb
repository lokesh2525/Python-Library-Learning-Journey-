Line Plot:

```python
import matplotlib.pyplot as plt

#sample data
x=[1,2,3,4,5,6]
y=[12,21,34,5,6,7]

# Create a line plot
plt.plot(x, y)

# Add a title and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

Scatter Plot:

```python
import matplotlib.pyplot as plt

#sample data
x=[1,2,3,4,5,6]
y=[5,6,7,9,5,4]

#create a scatter plot
plt.scatter(x, y, color='red')

# Add a title and labels
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```

Bar Chart:

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 5]

# Create a bar chart
plt.bar(categories, values, color='blue')

# Add a title and labels
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

# Display the plot
plt.show()
```

Histogram:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)

# Create a histogram
plt.hist(data,bins=30, color='green', alpha=0.7)

# Add a title and labels
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

pie plot:

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=['c','c++','java','python']
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.1)

# Add a title and labels
plt.title("Pie chart")
plt.xlabel("Value")
plt.ylabel("Name")

plt.show()
```

Basic Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.normal(0, 1, 100)

# Create a box plot
plt.boxplot(data)

# Add title and labels
plt.title('Box Plot of Random Data')
plt.ylabel('Values')

# Display the plot
plt.show()
```

 Box Plot with Multiple Data Sets

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for three different groups
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 2, 100)
data3 = np.random.normal(2, 1.5, 100)

# Combine data into a list
data = [data1, data2, data3]

# Create a box plot
plt.boxplot(data, patch_artist=True, labels=['Group 1', 'Group 2', 'Group 3'])

# Customize colors
colors = ['lightblue', 'lightgreen', 'lightpink']
for patch, color in zip(plt.boxplot(data, patch_artist=True)['boxes'], colors):
    patch.set_facecolor(color)

# Add title and labels
plt.title('Box Plot of Multiple Data Sets')
plt.xlabel('Groups')
plt.ylabel('Values')

# Display the plot
plt.show()
```

Box Plot with Outliers

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data with outliers
data = np.concatenate([np.random.normal(0, 1, 100), np.random.normal(10, 1, 5)])

# Create a box plot
plt.boxplot(data, showfliers=True)

# Add title and labels
plt.title('Box Plot with Outliers')
plt.ylabel('Values')

# Display the plot
plt.show()
```
